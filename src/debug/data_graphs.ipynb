```python
import matplotlib.pyplot as plt
import json

data_file = 'id_w_data.txt'
f = open(data_file)
big_data = json.loads(f.read())
big_data.keys()
```

```python
for key in big_data.keys():
    plt.plot([o for o in range(0, len(big_data[key]))], big_data[key], label=key)
# plt.legend()
plt.savefig("data_graphed.png")
plt.show()
```
